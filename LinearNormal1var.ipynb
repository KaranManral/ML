```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
diabetes_X,diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
diabetes_X = diabetes_X[:,np.newaxis,2]
diabetes_X
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
bias=np.ones((diabetes_X_train.size,1))
bias
```

```python
x_new = np.append(bias,diabetes_X_train,axis=1)
x_new
```

```python
x_new_transpose = x_new.transpose()
x_new_transpose
```

```python
x_new_transpose_dot_x_new = x_new_transpose.dot(x_new)
x_new_transpose_dot_x_new
```

```python
x_new_transpose_dot_x_new_inverse = np.linalg.inv(x_new_transpose_dot_x_new)
x_new_transpose_dot_x_new_inverse
```

```python
temp1 = x_new_transpose.dot(diabetes_y_train)
temp1
```

```python
theta = x_new_transpose_dot_x_new_inverse.dot(temp1)
theta
```

```python
Intercept = theta[0]
Slope =  theta[1]
print("Intercept: ",Intercept)
print("Slope: ",Slope)
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test)
plt.plot(diabetes_X_test,Slope*diabetes_X_test+Intercept,color='red')
```
