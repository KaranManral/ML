```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
diabetes_X,diabetes_y = datasets.load_diabetes(return_X_y=True)
diabetes_X
```

```python
diabetes_y
```

```python
diabetes_X = diabetes_X[:,np.newaxis,2]
diabetes_X
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
def hypothesis(x, w, b):
  y_pred = w*x+b
  return y_pred;
```

```python
def cost_error(y, y_pred):
  cost = (1/(2*len(y)))*np.sum((y_pred-y)**2)
  return cost
```

```python
def gradient_descent(x,y,w,b,itr,learning_rate,stop_value):
  cost=[]
  prev_cost=None
  weights=[]
  n=len(x)

  for i in range(itr):
    y_pred=hypothesis(x,w,b)
    current_cost=cost_error(y,y_pred)

    if prev_cost and abs(current_cost-prev_cost)<=stop_value:
      break

    prev_cost=current_cost

    temp_w=w-(learning_rate*(1/n)*np.sum((y_pred-y)*x))
    temp_b=b-(learning_rate*(1/n)*np.sum(y_pred-y))
    w=temp_w
    b=temp_b

    cost.append(current_cost)
    weights.append(w)

    print("Iteration :",i+1," Cost:",current_cost," Slope:",w," Intercept:",b)

  plt.plot(weights,cost)
  plt.scatter(weights,cost,marker='o',color='g')

  return w,b
```

```python
m,c=gradient_descent(diabetes_X_train,diabetes_y_train,w=0,b=0,itr=100,learning_rate=0.0001,stop_value=0.000001)
```

```python
print("Slope is ",m)
print("Intercept is ",c)

print("Equation is y=",m,"x +",c)

diabetes_y_pred = (m * diabetes_X_test)+c
# print("YP ",diabetes_y_pred)
# print("Cost is ",cost_error(diabetes_y_test,diabetes_y_pred))
plt.scatter(diabetes_X_test,diabetes_y_test)
plt.plot(diabetes_X_test,diabetes_y_pred,color='g')
plt.show()
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

print("YP",diabetes_y_pred)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

# plt.xticks(())
# plt.yticks(())

plt.show()
```
