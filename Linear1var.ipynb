```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def predict(x,y,w,b):
  y_pred=w*x+b
  return y_pred
```

```python
def cost_function(x,y,y_pred):
  n=len(x)
  cost_val=(1/2*n)*sum((y_pred-y)**2)
  return cost_val
```

```python
# x=np.array([1,2,3,4,5])
# y=np.array([3,6,7,12,6])

x = np.array([i for i in range(11)])
y = np.array([2*i for i in range(11)])
```

```python
def gradientDescent(x,y,learning_rate,iterations):
  w,b=0,0
  n=len(x)
  cost_error=[]
  weights=[]
  for i in range(iterations):
    # y_predicted=w*x+b
    y_predicted=predict(x,y,w,b)


    d_w=1/n*sum((y_predicted-y)*x)
    d_b=1/n*sum(y_predicted-y)

    # cost=(1/2*n)*sum((y_predicted-y)**2)
    cost=cost_function(x,y,y_predicted)
    w=w-learning_rate*d_w
    b=w-learning_rate*d_b

    cost_error.append(cost)
    weights.append(w)
  return w,b,cost_error,weights
```

```python
print("X Values:",x)
print("Y Values:",y)
```

```python
y_predicted=predict(x,y,1,2)
print("Y Predicted Values:",y_predicted)
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,y_predicted)
```

```python
w,b,cost_error,weights=gradientDescent(x,y,0.0001,iterations=1000)
print("Weight:",w)
print("Bias:",b)
print("Cost Error:",cost_error)
print("weights:",weights)
plt.plot(weights,cost_error)
```
